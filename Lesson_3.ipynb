1) Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
x = np.linspace(-5, 4, 201)
plt.plot(x, (1 - (1 - np.exp(x))/x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,5) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 - 1, np.exp(x) - x*(1-y) - 1)

x1, y1 =  fsolve(equations, (-2, 1))
              
print (x1, y1)
```
